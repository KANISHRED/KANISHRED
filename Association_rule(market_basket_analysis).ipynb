```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('GroceryStoreDataSet.csv', names = ['products'], sep = ',')
df.head()
```

```python
data = df.products.str.split('\s*,\s*', expand=True).stack().str.get_dummies().sum(level=0)
```

```python
data
```

```python
df_support = apriori(data,min_support=0.2, max_len =2,use_colnames = True)
#association_results=list(association_rules)
```

```python
df_support
```

```python
df_association_rules = association_rules(df_support, metric = "confidence", min_threshold = 0.6)
```
